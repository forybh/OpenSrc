## 광진구와 인구 분포 가장 다른곳과 가장 비슷한 곳 구하기
## 그래프로 나타내기

### 가장 다른 곳 : 송파구
### 가장 비슷한 곳 : 동대문구

```python
import csv
import matplotlib.pyplot as plt
import numpy as np


f = open('population.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
all_data = []
all_gu = []
print(all_data)
for row in data:
    temp = list(map(lambda x: "".join(x.split(',')),row))
    all_data.append(list(map(int, temp[3:])))
    all_gu.append(row[0])
np_data = np.array(all_data)

del all_gu[4]
del all_data[4]
gwang = np_data[4]

except_gwang = np.delete(np_data, 5, axis = 0)


np_operate = np.sum(np.subtract(except_gwang,gwang)**2,axis = 1)

max_index = np.argmin(np_operate)
min_index = np.argmax(np_operate)
print(all_gu[max_index], all_gu[min_index])


plt.style.use("ggplot")
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 10
plt.plot(all_data[max_index], color = 'b', label = all_gu[max_index])
plt.plot(all_data[min_index], color = 'r', label = all_gu[min_index])
plt.plot(gwang, color = 'g', label = "광진구")
plt.legend(loc='upper right')
plt.show()
```
