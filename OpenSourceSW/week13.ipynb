# 실습 1
## 8월 22일 최고/ 최저 기온 그래프로 나타내기


```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

hot = []
cold = []
for row in data:
    if row[-1] != '' :
        if row[0].split('.')[1] == '8' and row[0].split('.')[2] == '22':
            hot.append(float(row[-1]))
            cold.append(float(row[-2]))
print(hot, cold)
x = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(x,hot,'r',label='hot')
plt.plot(x,cold,'c',label='cold')
plt.title('8/22 hot/cold')
plt.legend(loc='upper right')

plt.show()
f.close()
```

# 실습2
## 1월 8월 최고기온 막대그래프로 나타내기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

hot1 = []
hot8 = []
for row in data:
    if row[-1] != '' :
        if row[0].split('.')[1] == '1':
            hot1.append(float(row[-1]))
        if row[0].split('.')[1] == '8':
            hot8.append(float(row[-1]))
plt.hist(hot1, bins=100, color='b', label = 'January')
plt.hist(hot8, bins=100, color='r', label = 'August')
plt.title('January, August highest[maximum] temperature')
plt.legend(loc='upper right')
plt.show()
```
