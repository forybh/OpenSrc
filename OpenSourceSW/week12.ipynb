# 실습 1
## 일교차 가장 큰 날
## 최저 기온이 가장 낮은 날
## 최고 기온이 가장 높은 날


```python
import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
max_d, max_l, max_h = -999, 999, -999
d_date, l_date, h_date = "", "", ""

for row in data:
    if max_d < abs(float(row[4])-float(row[3])):
        d_date = row[0]
        max_d = abs(float(row[4])-float(row[3]))
    if max_l > float(row[3]):
        l_date = row[0]
        max_l = float(row[3])
    if max_h < float(row[4]):
        h_date = row[0]
        max_h = float(row[4])
f.close()
print("일교차가 가장 큰 날")
print(d_date)
print("최저 기온이 가장 낮은 날")
print(l_date)
print("최고 기온이 가장 높은 날")
print(h_date)
```

# 실습 2
## 호선별 이용객
## 가장 많은 역 3순위
## 가장 적은 역 3순위

```python
import csv
f = open('subwaypassenger.csv')
data = csv.reader(f)
next(data)
station = "1호선"
stations = []
passengers = []

for row in data:
    if station is None:
        break     
    if row[1] == station:
        stations.append(row[3])
        passengers.append(int(row[4].replace(",", "")))
    else:
        print(station)
        print("이용객이 가장 많은 역 순위")
        for i in range(3):
            index = passengers.index(max(passengers))
            max_p = passengers.pop(index)
            max_s = stations.pop(index)
            print(f"{i+1}. {max_s}({max_p})")
        print("이용객이 가장 적은 역 순위")
        for i in range(3):
            if len(passengers)==0:
                print(print(f"{i+1}. {max_s}({max_p})"))
            else :
                index = passengers.index(min(passengers))
                min_p = passengers.pop(index)
                min_s = stations.pop(index)
                print(f"{i+1}. {min_s}({min_p})")
        station = row[1]
        stations.clear()
        passengers.clear()


f.close()
```
